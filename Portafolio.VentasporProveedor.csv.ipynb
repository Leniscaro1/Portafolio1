```python
VentaGorras_Norte =[23, 10, 9, 5, 7, 0]
x=range(len(VentaGorras_Norte))
x
```

```python
plt.bar(x, VentaGorras_Norte, color ='red')

ax= plt.subplot()
ax.set_xticks(x)
ax.set_xticklabels(['Ashton', 'Snodgrass', 'Bond','Delaney','Stanford','Gee'],rotation=30)


ax=plt.title("Ventas de Gorras en la región norte durante el mes de Marzo")
plt.show()
```

```python
# Construimos un gráfico que resume todo el proceso anterior

explode =[0, 0, 0, 0.1, 0, 0]
explode1=[0, 0, 0, 0, 0.1, 0]


plt.subplot(2,2,1)
Ganancia1 = [20, 29, 20, 30, 16, 23]
plt.title("Proveedores de la región Este")
plt.pie(Ganancia1, explode=explode, autopct='%.0f%%', labels=['Stanford','Delaney','Gee','Bond','Ashton','Snodgrass'] )


plt.subplot(2,2,2)
plt.subplots_adjust(wspace =0.5, hspace = 0.5)
Ganancia2 = [11, 18, 12, 24, 23, 12]
plt.title("Proveedores de la región Norte")
plt.pie(Ganancia2, explode=explode, autopct='%.0f%%', labels=['Stanford','Delaney','Gee','Bond','Ashton','Snodgrass'] )


plt.subplot(2,2,3)
Ganancia3 = [17, 9, 20, 20, 22 ,23]
plt.title("Proveedores de la región Oeste")
plt.pie(Ganancia3, explode=explode1, autopct='%.0f%%', labels=['Stanford','Delaney','Gee','Bond','Ashton','Snodgrass'] )


plt.subplot(2,2,4)
plt.subplots_adjust(wspace =0.5, hspace = 0.5)
Ganancia4 = [21, 20, 18, 21, 23, 17]
plt.title("Proveedores de la región Sur")
plt.pie(Ganancia4, explode=explode1, autopct='%.0f%%', labels=['Stanford','Delaney','Gee','Bond','Ashton','Snodgrass'] )


plt.show()
```

Bond es la mejor proveedora de la región Este y Norte, mientras Ashton es el mejor proveedor de la región Oeste y Sur
